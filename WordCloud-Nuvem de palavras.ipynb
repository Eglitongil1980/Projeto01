```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
df = pd.read_csv("listings.csv")
```

```python
df.shape
```

```python
df
```

```python
name = df.dropna(subset=['name'], axis = 0) ['name']
```

```python
display(name.iloc[1500])
```

```python
all_name = "".join(s for s in name)
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["da", "meu", "em", "você", "de", "ao", "os"])
```

```python
wordcloud = WordCloud(stopwords=stopwords,
                     background_color="black",
                     width=1600, height=800).generate(all_name)
```

```python
fig, ax = plt.subplots(figsize=(15,9))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
```

```python
wordcloud.to_file("airbnb_name_wordcloud.png")
```

```python
rio_mask = np.array(Image.open("cristoRedentor.jpg"))
```

```python
wordcloud = WordCloud(stopwords=stopwords,
                     background_color="white",
                     width=1000, height=1000, max_words=2000,
                     mask=rio_mask, max_font_size=200,
                     min_font_size=1).generate(all_name)
```

```python
fig, ax = plt.subplots(figsize=(15,9))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
```
